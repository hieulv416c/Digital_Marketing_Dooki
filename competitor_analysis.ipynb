# PTICH TEXT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import string
from wordcloud import WordCloud

# Đọc dữ liệu
df = df_time.copy()

# Xử lý NaN
df["text"] = df["text"].fillna("")
df["user/name"] = df["user/name"].fillna("Unknown")  # Xử lý trường hợp thiếu tên thương hiệu

# Danh sách stopwords tiếng Việt (có thể mở rộng)
vietnamese_stopwords = set([
    "là", "và", "có", "của", "cho", "với", "được", "nhưng", "này", "các",
    "một", "về", "không", "đến", "tại", "thì", "bạn", "chúng", "tôi", "cũng"
])

# HÀM XỬ LÝ VĂN BẢN
def preprocess_text(text):
    text = text.lower().translate(str.maketrans("", "", string.punctuation))  # Chuyển thành chữ thường & bỏ dấu câu
    words = text.split()
    words = [word for word in words if word not in vietnamese_stopwords]  # Loại bỏ stopwords
    return words

df["clean_text"] = df["text"].apply(preprocess_text)

# Tạo danh sách các thương hiệu
brands = df["user/name"].unique()

#1. **Phân tích từ khóa phổ biến theo từng thương hiệu**
for brand in brands:
    df_brand = df[df["user/name"] == brand]

    # Đếm từ khóa
    all_words = [word for sublist in df_brand["clean_text"] for word in sublist]
    word_freq = Counter(all_words).most_common(10)  # Lấy 10 từ phổ biến nhất

    if word_freq:  # Chỉ vẽ nếu có dữ liệu
        words, counts = zip(*word_freq)
        plt.figure(figsize=(10, 5))
        sns.barplot(x=list(words), y=list(counts))
        plt.xticks(rotation=45)
        plt.title(f"10 từ phổ biến nhất - {brand}")
        plt.ylabel("Tần suất")
        plt.show()

# 2. **Vẽ Word Cloud theo thương hiệu**
for brand in brands:
    df_brand = df[df["user/name"] == brand]

    # Gom tất cả các từ lại
    all_words = [word for sublist in df_brand["clean_text"] for word in sublist]

    if all_words:  # Kiểm tra xem có dữ liệu không
        wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(all_words))
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.title(f"Word Cloud - {brand}")
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Trích xuất ngày trong tuần và giờ
df_time["day_of_week"] = df_time["timestamp"].dt.day_name()
df_time["hour"] = df_time["timestamp"].dt.hour

# Ánh xạ thứ tự ngày trong tuần (để hiển thị đúng thứ tự)
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df_time["day_of_week"] = pd.Categorical(df_time["day_of_week"], categories=days_order, ordered=True)

# Tính tổng số bài đăng & tương tác theo giờ/ngày
agg_df = df_time.groupby(["day_of_week", "hour"]).agg(
    total_posts=("user/name", "count"),
    total_engagement=("total_engagement", "sum")
).reset_index()

# Vẽ biểu đồ bong bóng
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    data=agg_df,
    x="day_of_week", y="hour",
    size="total_posts", hue="total_engagement",
    sizes=(10, 500), palette="RdYlGn", edgecolor="black", alpha=0.8
)

# Tuỳ chỉnh hiển thị
plt.xlabel("")
plt.ylabel("Giờ trong ngày")
plt.title("Lịch đăng bài và mức độ tương tác")
plt.xticks(rotation=45)
plt.legend(
    title="Tương tác (màu) & Số bài đăng (kích thước)",
    bbox_to_anchor=(1.05, 1), loc="upper left"
)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Chuyển đổi cột 'month' thành kiểu datetime và lấy năm
df["year"] = pd.to_datetime(df["month"]).dt.year

# Định nghĩa màu cho các loại sentiment (chú ý chữ hoa đầu)
sentiment_colors = {"Positive": "#66c2a5", "Negative": "#fc8d62", "Neutral": "#8da0cb"}

# Lấy danh sách các thương hiệu
brands = df["user/name"].unique()

# Vẽ biểu đồ cho từng thương hiệu
for brand in brands:
    fig, ax = plt.subplots(figsize=(10, 5))
    brand_data = df[df["user/name"] == brand]

    # Đếm số lượng sentiment theo năm
    sentiment_counts = brand_data.groupby(["year", "sentiment_name"]).size().unstack().fillna(0)

    # Kiểm tra danh sách sentiment thực tế có trong dữ liệu
    available_sentiments = sentiment_counts.columns.tolist()
    sentiment_colors_filtered = [sentiment_colors[s] for s in available_sentiments if s in sentiment_colors]

    # Vẽ biểu đồ cột nhóm
    bars = sentiment_counts.plot(kind="bar", width=0.8, color=sentiment_colors_filtered, ax=ax)

    # Thêm số trên mỗi cột
    for container in bars.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)

    ax.set_title(brand, fontsize=12, fontweight="bold")
    ax.set_xlabel("Year")
    ax.set_ylabel("Count")
    ax.legend(title="Sentiment")

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lấy danh sách thương hiệu
brands = df_time["user/name"].unique()

# Lặp qua từng thương hiệu
for brand in brands:
    # Lọc dữ liệu theo thương hiệu
    df_brand = df_time[df_time["user/name"] == brand]

    # 1. Tạo dữ liệu gộp: số lượng bài post và tổng tương tác theo giờ
    hourly_data = (
        df_brand.groupby("hour")
                .agg({
                    "user/id": "count",            # Đếm số bài viết
                    "total_engagement": "sum"     # Tổng lượt tương tác
                })
                .reset_index()
                .rename(columns={"user/id": "num_posts"})
    )

    # 2. Vẽ biểu đồ kết hợp
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Biểu đồ cột: số lượng bài post
    ax1.bar(hourly_data["hour"], hourly_data["num_posts"], color='skyblue')
    ax1.set_xlabel("Giờ đăng bài")
    ax1.set_ylabel("Số lượng bài đăng", color='blue')

    # Tạo trục tung thứ hai cho tổng lượt tương tác
    ax2 = ax1.twinx()
    ax2.plot(hourly_data["hour"], hourly_data["total_engagement"], color='red', marker='o')
    ax2.set_ylabel("Tổng lượt tương tác", color='red')

    plt.title(f"Số lượng bài post (cột) và Tổng lượt tương tác (đường) theo giờ - {brand}")
    plt.show()
```
