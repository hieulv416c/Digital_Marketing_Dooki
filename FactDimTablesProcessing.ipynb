# Xử lý dữ liệu text và thêm 2 cột phân loại content và số lượng từ trong text.

```python
import re

def process_text_features(df):
    # --- 1. Đếm số lượng từ (Word Count) ---
    def count_words(text):
        t = str(text)
        if t.lower() == 'no content' or t.strip() == "":
            return 0
        return len(t.split())

    df['word_count'] = df['text'].apply(count_words)

    # --- 2. Phân loại nội dung (Content Category) ---
    def classify_content(text):
        t = str(text).lower()

        # Xử lý trường hợp không có nội dung chữ
        if t == 'no content' or t.strip() == "":
            return 'Media Only'

        # Nhóm 1: Khuyến mãi / Giá cả
        if any(w in t for w in ['khuyến mãi', 'ưu đãi', 'giảm', '139k', 'giá', 'voucher', 'buffet', 'chỉ từ', 'sale']):
            return 'Promotion'

        # Nhóm 2: Tương tác / Game
        elif any(w in t for w in ['minigame', 'tặng', 'quà', 'may mắn', 'thử thách', 'tag', 'comment', 'nhận ngay']):
            return 'Engagement'

        # Nhóm 3: Món ăn / Review (Đặc thù F&B)
        elif any(w in t for w in ['ngon', 'vị', 'gà', 'lẩu', 'tokbokki', 'phô mai', 'món mới', 'menu', 'thưởng thức']):
            return 'Food/Product'

        # Nhóm 4: Tin tức / Khai trương
        elif any(w in t for w in ['khai trương', 'chi nhánh', 'địa chỉ', 'tầng', 'thông báo', 'tuyển dụng']):
            return 'News/Branch'

        # Còn lại: Thương hiệu / Khác
        else:
            return 'Branding/Others'

    df['content_category'] = df['text'].apply(classify_content)
    return df
```

```python
# Áp dụng cho từng bảng của bạn
df_doki_cleaned = process_text_features(df_doki_cleaned)
df_spicy_cleaned = process_text_features(df_spicy_cleaned)
df_joseon = process_text_features(df_joseon)
df_joppii = process_text_features(df_joppii)

print("Đã thêm cột word_count và content_category thành công!")
```

# Tạo Dimtime

```python
import pandas as pd

# 1. Tạo dãy thời gian từ 2018 đến giữa 2025 (tần suất theo Giờ 'H')
time_rng = pd.date_range(start='2018-01-01', end='2025-06-30 23:00:00', freq='H')

# 2. Tạo DataFrame
dim_time = pd.DataFrame(time_rng, columns=['timestamp'])

# 3. Trích xuất các cột thời gian cơ bản
dim_time['full_date'] = dim_time['timestamp'].dt.date
dim_time['year'] = dim_time['timestamp'].dt.year
dim_time['month'] = dim_time['timestamp'].dt.month
dim_time['day'] = dim_time['timestamp'].dt.day
dim_time['hour'] = dim_time['timestamp'].dt.hour
dim_time['day_of_week'] = dim_time['timestamp'].dt.day_name()

# 4. THÊM CỘT QUÝ (Quarter)
# Kết quả sẽ là 1, 2, 3, 4
dim_time['quarter'] = dim_time['timestamp'].dt.quarter

# 5. THÊM CỘT TYPE DAY (Weekday/Weekend)
# dt.dayofweek trả về 0-4 là từ Thứ 2 đến Thứ 6, 5-6 là Thứ 7 & CN
dim_time['type_day'] = dim_time['timestamp'].dt.dayofweek.apply(
    lambda x: 'Weekend' if x >= 5 else 'Weekday'
)

# 6. Phân loại "Buổi" (Daypart Analysis)
def phan_loai_buoi(h):
    if 5 <= h < 11: return 'Morning'
    elif 11 <= h < 14: return 'Noon'
    elif 14 <= h < 17: return 'Afternoon'
    elif 17 <= h < 22: return 'Evening'
    else: return 'Night'

dim_time['day_part'] = dim_time['hour'].apply(phan_loai_buoi)

# Hiển thị thử 5 dòng ngẫu nhiên để kiểm tra
print(dim_time.sample(5))
```

# Tạo Fact/Dim Table


```python
import pandas as pd

# 1. Đồng nhất tên cột isVideo
df_joppii = df_joppii.rename(columns={'isvideo': 'isVideo'})

# 2. Danh sách cột cho bảng Fact (Đã LOẠI BỎ cột 'text')
fact_columns = [
    'postId', 'pageName', 'timestamp', 'likes', 'comments',
    'shares', 'topReactionsCount', 'isVideo', 'viewsCount',
    'word_count', 'content_category'
]

# 3. Gộp 4 bảng thành Fact_Posts
fact_posts = pd.concat([
    df_doki_cleaned[fact_columns],
    df_spicy_cleaned[fact_columns],
    df_joseon[fact_columns],
    df_joppii[fact_columns]
], ignore_index=True)

# --- 4. XỬ LÝ TIMESTAMP (Hàm fix cũ của chúng ta) ---
def fix_timestamp(ts):
    try:
        return pd.to_datetime(ts)
    except:
        try:
            val = float(ts)
            if val > 9999999999:
                return pd.to_datetime(val, unit='ms')
            return pd.to_datetime(val, unit='s')
        except:
            return pd.NaT

fact_posts['timestamp'] = fact_posts['timestamp'].apply(fix_timestamp)
fact_posts = fact_posts.dropna(subset=['timestamp'])

# Tạo khóa nối với bảng Dim_Time và chuẩn hóa isVideo
fact_posts['timestamp_link'] = fact_posts['timestamp'].dt.floor('H')
fact_posts['isVideo'] = pd.to_numeric(fact_posts['isVideo'], errors='coerce').fillna(0).astype(int)

# --- 5. XUẤT CÁC BẢNG RA CSV CHO POWER BI ---

# File 1: Fact_Posts (Gọn nhẹ, không có text)
fact_posts.to_csv('Fact_Posts.csv', index=False, encoding='utf-8-sig')

# File 2: Dim_Time (Bảng thời gian bạn đã tạo)
dim_time.to_csv('Dim_Time.csv', index=False, encoding='utf-8-sig')

# File 3: Dim_Pages (Phân loại Brand)
dim_pages = pd.DataFrame({'pageName': fact_posts['pageName'].unique()})
dim_pages['Brand_Group'] = dim_pages['pageName'].apply(
    lambda x: 'Dookki' if 'doki' in str(x).lower() or 'dookki' in str(x).lower() else 'Competitor'
)
dim_pages.to_csv('Dim_Pages.csv', index=False, encoding='utf-8-sig')

# File 4: Dim_Dookki_Details (Giữ lại Comment_Text và Sentiment cho Dookki)
dim_dookki_details = df_doki_cleaned[
    ['commentId', 'postId', 'Comment_Text', 'Sentiment_Score', 'date', 'likesCount']
].dropna(subset=['commentId'])
dim_dookki_details.to_csv('Dim_Dookki_Details.csv', index=False, encoding='utf-8-sig')

print("✅ Đã xử lý xong! Bảng Fact giờ đây chỉ chứa dữ liệu số và các category.")
display(fact_posts.head())
```

```python
# --- 2. TẠO BẢNG DIM_DOOKKI_DETAILS (Bảng chi tiết cho Dookki) ---

# Tách riêng phần comment và sentiment của Dookki để tránh làm loãng bảng Fact
dim_dookki_details = df_doki_cleaned[
    ['commentId', 'postId', 'Comment_Text', 'Sentiment_Score', 'date', 'likesCount']
].dropna(subset=['commentId']).copy()
```

```python
# --- 3. TẠO BẢNG DIM_PAGES (Thông tin thương hiệu) ---

dim_pages = pd.DataFrame({
    'pageName': fact_posts['pageName'].unique()
})

# Phân loại Brand để báo cáo (Target vs Competitor)
dim_pages['brand_label'] = dim_pages['pageName'].apply(
    lambda x: 'Dookki' if 'doki' in str(x).lower() or 'dookki' in str(x).lower() else x
)
dim_pages['is_target'] = dim_pages['brand_label'].apply(lambda x: 1 if x == 'Dookki' else 0)
```
